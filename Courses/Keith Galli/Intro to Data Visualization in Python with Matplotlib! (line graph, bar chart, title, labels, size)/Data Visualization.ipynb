# Data Visualization

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Basic graph
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.figure(figsize = (9, 4))

plt.title('Title\nSubtitle', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 10})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

plt.plot(x, y, label = '2x', color = 'green', linewidth = 2, marker = 'o', markersize = 10, markeredgecolor = 'black', linestyle = '--')

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:4], x2[:4] ** 2, linestyle = '--', label = 'X^2', color = 'blue')
plt.plot(x2[3:], x2[3:] ** 2, label = 'X^2', color = 'blue')

plt.legend()

plt.savefig('graph.png', dpi = 300)

plt.show()
```
