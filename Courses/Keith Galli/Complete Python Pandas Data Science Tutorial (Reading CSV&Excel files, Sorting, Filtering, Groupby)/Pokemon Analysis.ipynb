```python
# Imports
import pandas as pd
```

```python
# load Data

# csv
df = pd.read_csv('datasets/pokemon_data.csv')
df.head()

# xlsx
# df_xlsx = pd.read_excel('datasets/pokemon_data.xlsx')
# df_xlsx.head()

# txt
# df_txt = pd.read_csv('datasets/pokemon_data.txt', delimiter = '\t')
# df_txt.head()
```

```python
# Get columns
df.columns

# Get by column
df[['Name', 'HP', 'Attack']]

# Get by row
df.iloc[1:3]

# Get by row & column
df.iloc[2, 1]

# Iteration
# for index, row in df.iterrows():
#     print(index, row['Name'])
    
# Get by condition
df.loc[df['Type 1'] == 'Grass']
```

```python
# Describing data
df.describe()

# Sorting data
df.sort_values(by = ['HP', 'Type 1'], ascending = [True, False])
```

```python
# Manipulating data

# df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df['Total'] = df.iloc[:, 4:10].sum(axis = 1)

df.head(3)
```

```python
# Drop columns
df.drop(columns = ['Total'], inplace = True)
df.head(3)
```

```python
# Reposition columns
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:-1]]
df.head(3)
```

```python
# Save the dataframe

# As csv
df.to_csv('datasets/modified_pokemon_data.csv', index = False)

# Ax Xlsx
df.to_excel('datasets/modified_pokemon_data.xlsx', index = False)

# As txt
df.to_csv('datasets/modified_pokemon_data.txt', index = False, sep = '\t')
```
