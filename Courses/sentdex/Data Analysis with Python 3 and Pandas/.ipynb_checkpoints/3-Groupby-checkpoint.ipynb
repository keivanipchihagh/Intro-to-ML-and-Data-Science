```python
import pandas as pd

df = pd.read_csv('datasets/minimum_wage_data.csv')

df.head()
```

```python
gb = df.groupby('State')
gb.get_group('Alabama').set_index('Year').head()
```

```python
act_min_wage = pd.DataFrame()

for name, group in df.groupby('State'):
    
    if act_min_wage.empty:
        act_min_wage = group.set_index('Year')[['Federal.Minimum.Wage.2020.Dollars']])
    else:
        act_min_wage = act_min_wage.join(group.set_index('Year')[['Federal.Minimum.Wage.2020.Dollars']])
        
act_min_wage.head()
```
