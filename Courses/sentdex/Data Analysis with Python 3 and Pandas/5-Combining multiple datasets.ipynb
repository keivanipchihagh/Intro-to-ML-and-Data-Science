```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

unemp_county = pd.read_csv('datasets/unemployment_rate_by_county.csv')
unemp_county.head()
```

```python
df = pd.read_csv('datasets/minimum_wage_data.csv')
df = df.copy()

act_min_wage = pd.DataFrame()

for name, group in df.groupby('State'):
    
    if act_min_wage.empty:
        act_min_wage = group.set_index('Year')[['Effective.Minimum.Wage.2020.Dollars']].rename(columns = {'Effective.Minimum.Wage.2020.Dollars': name})
    else:
        act_min_wage = act_min_wage.join(group.set_index('Year')[['Effective.Minimum.Wage.2020.Dollars']].rename(columns = {'Effective.Minimum.Wage.2020.Dollars': name}))
        
act_min_wage.head()
```

```python
act_min_wage = act_min_wage.replace(0, np.NaN).dropna(axis = 1)
act_min_wage.head()
```

```python
def get_min_wage(year, state):
    try:
        return act_min_wage.loc[year][state]
    except:
        return np.Nan
    
get_min_wage(2012, 'Colorado')
```

```python
unemp_county['wage'] = map(get_min_wage, unemp_county['Year'], unemp_county['State'])
```
